```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
superbowl_df = pd.read_csv ('/Users/lhabersham/code/ETL_Project/Data/superbowl.csv')
superbowl_df.sample(10)
```

```python
ad_df = pd.read_csv ('/Users/lhabersham/code/ETL_Project/Data/superbowl-ads.csv')
ad_df = ad_df.rename(columns={'Year':'year',
                              'Product Type':'product_type',
                              'Product/Title':'product_title',
                              'Plot/Notes':'plot_notes'}
                    )
ad_df.sample(10)
```

```python
ad_df["ad_brand"] = ad_df['product_title'].str.partition('"')[0]
ad_df['year']= ad_df['year'].astype(int)
ad_df
```

```python
superbowl_df['sb_year'] = superbowl_df['sb_date'].str[-4:]
superbowl_df = superbowl_df.rename(columns={'sb_year':'year'})
ad_df['year']= ad_df['year'].astype(int)
superbowl_df
```

```python
#separate SB Numeral
superbowl_df['sb_numeral']=superbowl_df['sb'].str.partition(" ")[0]
superbowl_df
```

```python
superbowl_df["year"] = superbowl_df["year"].astype(int) 
```

```python
superbowlfacts = pd.merge(ad_df, superbowl_df, on='year')
superbowlfacts
```

```python
rds_connection_string = "postgres:3T#thu#3@localhost:5432/Superbowl_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
ad_df.to_sql(name='superbowlads', con=engine, if_exists='replace', index=False)
```

```python
superbowl_df.to_sql(name='superbowl', con=engine, if_exists='replace', index=False)
```

```python
superbowlfacts.to_sql(name='superbowlfacts', con=engine, if_exists='replace', index=False)
```
